```python
import pandas as pd
```

```python
student_details={'stud_id':[1,2,3,4,5],
                'stud_name':['A','B','C','D','E'],
                'Region':['N','S','E','W','E']}
```

```python
sales_details={'st_id':[1,2,3,4,6],
              'Sub':['Py','R','Tab','SQL','Py'],
              'sales':[250,125,200,200,250]}
```

```python
student_details=pd.DataFrame(student_details)
```

```python
student_details
```

```python
sales_details=pd.DataFrame(sales_details)
```

```python
sales_details
```

```python
pd.merge(student_details,sales_details,left_on='stud_id',right_on='st_id',how='inner')
```

```python
pd.merge(student_details,sales_details,left_on='stud_id',right_on='st_id',how='left')
```

```python
pd.merge(student_details,sales_details,left_on='stud_id',right_on='st_id',how='right')
```

```python
pd.merge(student_details,sales_details, left_on='stud_id',right_on='st_id',how='outer')
```
